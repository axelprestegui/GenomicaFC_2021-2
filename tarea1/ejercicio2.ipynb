2. En el archivo promotores.txt se encuentra la lista de secuencias tomadas
del genoma de Vitis vinifera y cada una de las secuencias puede que tenga
alguno de las diferentes formas en las que se ha encontrado el promotor
GATA: {AGATAG, TGATAG, AGATAA, TGATAA}
Deseamos estudiar estas regiones en función del promotor GATA y por lo
tanto lo primero que deseamos es saber cuántas veces aparecen los promotores en cada región.
Tu respuesta deberá ser el conteo de la cantidad de veces que la región
i-ésima contiene al promotor GATA en cualquiera de sus formas. Ejemplo:
[3, 2, 1, ...]
Quiere decir que la primer secuencia contiene 3 instancias del promotor,
la segunda tiene 2, la tercera 1 y así en adelante.

```python
import re
# Nuestra expresión regular
exp =  r'[AT]GATA[GA]'
# Abrimos el archivo
with open('promotores.txt', 'r') as archivo:
    # Obtenemos todas las lineas del archivo
    lineas = archivo.read().splitlines()
    # Obtenemos todas las regiones
    regiones = []
    for linea in lineas:
        linea = linea.split()
        regiones.append(linea[1])
    # Para cada región, obtenemos el número de veces que aparece un promotor
    promotores_por_region = []
    for region in regiones:
        # Obtenemos todas las coincidencias
        coincidencias = re.finditer(exp, region)
        # Índice para contar cuántas ocurrencias hubo
        i = 0
        for m in coincidencias:
            # Por cada ocurrencia, sumamos uno
            i = i + 1
        # Agregamos a la lista de promotores_por_region la cantidad total de promotores encontrados en cada región
        promotores_por_region.append(i)
    # Finalmente imprimimos nuestra respuesta
    print(promotores_por_region)
```
